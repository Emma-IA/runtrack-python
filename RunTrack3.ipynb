## Job1.3

```python
data= open("data.txt","r")
list_of_words=[]
for line in data:
    words=line.split(" ")
    words=[''.join(e for e in word if e.isalnum()) for word in words]
    for word in words:
        if word:
            list_of_words.append(word)
print("nombre de mots avec doublons:",len(list_of_words))
list_no_double=list(set(list_of_words))
print("nombre de mots sans doublons:",len(list_no_double))

```

```python
import string
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

def count_alphabet():
    
    alphabet=list(string.ascii_lowercase)
    counter= pd.Series(np.zeros(len(alphabet), dtype=int), index = alphabet)
    for word in list_of_words:
        for letter in word:
            letter=str.lower(letter)
            if letter in counter.index:
                counter[letter]+=1
    return counter

counter=count_alphabet()
print(type(counter))
plt.bar(counter.index,100*counter.values/sum(counter.values))
plt.title("Pourcentage d'apparition des lettres dans data.txt")
```

```python

def begin_letter():
    alphabet=list(string.ascii_lowercase)
    counter= pd.Series(np.zeros(len(alphabet), dtype=int), index = alphabet)
    for word in list_of_words:
        letter=str.lower(word[0])
        if letter in counter.index:
            counter[letter]+=1
    return counter

counter=begin_letter()
plt.bar(counter.index,100*counter.values/sum(counter.values))
plt.title("Pourcentage d'apparition des lettres dans data.txt")
```

## Job13

```python

alphabet=list(string.ascii_lowercase)
def count_letter_after():
    counter=np.zeros((26,26))
    for word in list_of_words[:100]:
        for i in range(len(word)-1):
            if word[i] in alphabet and word[i+1] in alphabet:
                ind1=alphabet.index(word[i])
                ind2=alphabet.index(word[i+1])
                counter[ind1][ind2]+=1
    return counter


counter=count_letter_after()
for i in range(26):
    plt.plot(alphabet,counter[i]/26)
```
