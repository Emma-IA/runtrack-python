```python
def factorielle(nb):
    if nb==1:
        return nb
    else:
        return nb*factorielle(nb-1)
   
factorielle(4)
```

```python
def jeu(n):   
    if n<=3:
        return "Impossible"
    elif n==4:
        return [2, 4, 1, 3]
    else: 
        Liste=jeu(n-1)
        i=0
        if (n % 2) == 0:
            Liste.insert(0,n)
        else:
            Liste.append(n)
        return Liste
                                
# jeu(7)

def plateau(n):
    plateau=np.zeros((n,n), dtype=str)
    positions=jeu(n)
    for i in range(n):
        plateau[i][positions[i]-1]="X"
    return plateau
    
plateau(8)
```

```python

def compare_chars(char1, char2):
    n1=len(char1)
    n2=len(char2)
    if n1==n2:
        if char1==char2: 
            print("ok final")
            return 1
        else:
            print("false final")
            return 0
    else: 
        if char1[-1]==char2[-1]:
            print("ok direct")
            return compare_chars(char1[:-1],char2[:-1])
        elif char2[-1]=="*":
            i=2
            while char2[-i]=="*":
                i+=1
            if char2[-i]==char1[-1]:
                print("ok après étoile")
                return compare_chars(char1[:-1],char2[:-i])
            elif char2[-i] in char1[-1-i:-1]:
                indice= char1[-1-i:-1].index(char2[-i])
                print("ok étoile remplace lettre(s)")
                return compare_chars(char1[:indice],char2[:-i])
                
        else:
            print("false")
            return 0

compare_chars("rouge","ro**g*e")
        
```
